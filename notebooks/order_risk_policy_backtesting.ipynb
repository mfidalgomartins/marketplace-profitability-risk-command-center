# Order Risk Policy Backtesting

This notebook reviews threshold trade-offs for `order_risk_score` and simulates intervention impact using configurable efficacy and cost assumptions.

```python
import pandas as pd
import matplotlib.pyplot as plt

curve = pd.read_csv('../data/processed/backtesting_threshold_curve.csv')
impact = pd.read_csv('../data/processed/backtesting_action_impact.csv')
recommended = pd.read_csv('../data/processed/backtesting_recommended_policy.csv')
curve.head(), recommended
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(curve['threshold'], curve['precision'], marker='o', label='precision')
ax.plot(curve['threshold'], curve['recall'], marker='o', label='recall')
ax.plot(curve['threshold'], curve['review_rate'], marker='o', label='review_rate')
ax.set_title('Threshold Trade-off: Precision vs Recall vs Coverage')
ax.set_xlabel('Order Risk Threshold')
ax.set_ylabel('Rate')
ax.legend();
plt.show()
```

```python
pivot = impact.pivot(index='threshold', columns='efficacy_assumption', values='net_benefit')
fig, ax = plt.subplots(figsize=(10,5))
for c in pivot.columns:
    ax.plot(pivot.index, pivot[c] / 1_000_000, marker='o', label=f'efficacy={c:.2f}')
ax.set_title('Net Benefit Sensitivity by Efficacy Assumption')
ax.set_xlabel('Order Risk Threshold')
ax.set_ylabel('Net Benefit ($M)')
ax.legend();
plt.show()
```
